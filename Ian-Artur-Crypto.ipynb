```python
import matplotlib.pyplot as plt
```

```python
depo = 1239
print(f' = {depo}$')
print(f'    BTC = {round((depo * 0.7) * 0.8)}$')
print(f'    ETH = {round((depo * 0.7) * 0.2)}$')
print(f'   XLM = 23$,   ')
alts = depo - (round((depo * 0.7) * 0.8) + round((depo * 0.7) * 0.2) + 23)
print(f'    =  {depo - (round((depo * 0.7) * 0.8) + round((depo * 0.7) * 0.2) + 23)}$')
print(f'     (dot, atom, near, matic, sol, flow) = {round(alts * 0.6)}$')
print(f'     (mina, dydx, apt) = {round(alts * 0.4)}$')
print(f'    dot, atom, near, matic, sol, flow = {round(alts * 0.6) // 6}$')
print(f'    mina, dydx, apt = {round(alts * 0.4) // 3}$')
```

```python
labels = [' : 890$', ' : 209$', ' : 140$']
share = [890, 209, 140]
explode = (0.1, 0, 0)
plt.style.use('ggplot')
plt.title('   1239$')
plt.pie(x=share, explode=explode, labels=labels, autopct='%.2f%%', startangle=90)
plt.axis('equal')
plt.legend(loc='upper left')
plt.show()
```
